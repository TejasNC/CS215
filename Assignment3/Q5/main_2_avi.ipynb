```python
# Standardising the data set
def standardise(data):
	# Standardising the data
	data_mean = data.mean()
	data_std = data.std()
	data = (data - data_mean) / data_std
	return data, data_mean, data_std

train_X, train_mean, train_std = standardise(train.iloc[:, 13:16])
print(train_X)
train_Y = train.iloc[:, 16]
print(train_Y)
```

```python
def get_x(i, m, train_set):
    '''returns the x matrix for the polynomial regression'''
    # get the x column
    x = train_set.iloc[:, i].values
    x = x.flatten()
    #get a matrix where the ith row is x^i in numpy
    x = np.array([x**i for i in range(1,m+1)]).T
    # print("x shape: ", x.shape)
    return x

def get_other_x(i, m, train_set):
    '''returns all colums before the ith column in the train set'''
    x = train_set.iloc[:, :i].values

    return x

def get_y(j, train_set):
    '''returns the y column for the polynomial regression'''
    y = train_set.iloc[:, j].values
    # print("y shape: ", y.shape)
    return y

def get_weights(X, Y):
    '''returns the weights for the polynomial regression'''
    return np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
```

```python
def r2_score(X, Y, B):
    '''returns the r2 score'''
    return 1 - (np.sum((Y - X.dot(B))**2) / np.sum((Y - np.mean(Y))**2))

def ssr(X, Y, B):
    '''returns the sum of squared residuals'''
    return np.sum((Y - X.dot(B))**2)
```

```python
def k_fold_cross_validation(y, k, train_set):
    '''returns the k fold cross validation for all degrees'''
    k_fold = {}
    for i in range(1, 10):
        for a in range(1,10):
            for b in range(1,5):
                for c in range(1, 5):
                    # print(i,a,b)
                    ssr_list = []
                    for j in range(k):
                        # get the jth fold
                        # print(j)
                        test = train_set.iloc[j::k, :]
                        train = train_set.drop(train_set.index[j::k])
                        train_X0 = get_x(12, c, train)
                        train_X1 = get_x(13, i, train)
                        train_X2 = get_x(14, a, train)
                        train_X3 = get_x(15, b, train)
                        train_Xbefore = get_other_x(12, i, train)
                        print(train_Xbefore.shape,train_X0.shape, train_X1.shape, train_X2.shape, train_X3.shape)
                        # train_X = np.ones((train_X1.shape[0], 1))
                        train_X = np.concatenate((np.ones((train_X1.shape[0], 1)),train_Xbefore, train_X0, train_X1, train_X2, train_X3), axis=1)
                        # print(train_X)
                        train_Y = get_y(y, train)

                        B = get_weights(train_X, train_Y)

                        test_X0 = get_x(12, c, test)
                        test_X1 = get_x(13, i, test)
                        test_X2 = get_x(14, a, test)
                        test_X3 = get_x(15, b, test)
                        test_Xbefore = get_other_x(12, i, test)
                        test_X = np.concatenate((np.ones((test_X1.shape[0], 1)),test_Xbefore, test_X0, test_X1, test_X2, test_X3), axis=1)
                        test_Y = get_y(y, test)
                        # ssr = np.sum((test_Y - test_X.dot(B))**2)
                        ssr_list.append(ssr(test_X, test_Y, B))

                    k_fold[(i, a, b, c)] = np.mean(ssr_list)
                    # print("Degree: ", i, "SSR: ", np.mean(ssr_list))
    return k_fold

best_m_dict = k_fold_cross_validation(16, 10, train)

# for i in range(13, 16):
#     print(i)
#     best_m_list.append(np.argmin(k_fold_cross_validation(i, 16, 10, train)))
```

```python
def get_trainX(train, best_m):
    '''returns the train X matrix'''
    train_X0 = get_x(12, best_m[3], train)
    train_X1 = get_x(13, best_m[0], train)
    train_X2 = get_x(14, best_m[1], train)
    train_X3 = get_x(15, best_m[2], train)
    train_Xbefore = get_other_x(12, best_m[0], train)
    train_X = np.concatenate((np.ones((train_X1.shape[0], 1)),train_Xbefore, train_X0, train_X1, train_X2, train_X3), axis=1)
    return train_X

def get_trainY(train_set):
    '''returns the train Y matrix'''
    return get_y(16, train)
```
