Import the required libraries

```python
import matplotlib.pyplot as plt	
import numpy as np
import seaborn as sns
import pandas as pd
```

Load the dataset

```python
df = pd.read_csv("rows.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Types of issues
issue = df["Issue"].value_counts() # This returns the count of each type of issue with the issues sorted in descending order according to their counts
cumulative = issue.cumsum() / issue.sum() * 100
```

Plot the Pareto Chart of number of complaints v/s the types of customer complaints

```python
# plt.figure(figsize=(20, 5))
fig, ax = plt.subplots(figsize=(20, 5))
plt.title("Customer Complaints", fontsize=15)
ax.bar(issue.index, issue.values, color="C0", width=0.8)  # Increase the width to 0.8
plt.xlabel("Type of customer complaints", fontsize=12)
plt.ylabel("Number of complaints", fontsize=12)
plt.xticks(rotation=270, fontsize=8)
plt.yticks(fontsize=8)
ax2 = ax.twinx()
ax2.plot(issue.index, cumulative, color="C1", marker=".", ms=7)
plt.yticks(fontsize=8)
ax2.set_ylim(ymin=0)

plt.show()
```

We can see through this pareto chart that 80% of the total complaints can be solved by solving the first 20% of the different complaints that the customers had raised.
